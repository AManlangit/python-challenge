```python
from pathlib import Path
import csv
```

```python
total_months = 0
net_total_profit_losses = 0
previous_month_profit_loss = 0
monthly_changes = []
greatest_increase = ["", 0]
greatest_decrease = ["", 0]
```

```python
csv_path = Path("Resources/budget_data.csv")
```

```python
with open(csv_path, newline="") as csvfile:
    csvreader = csv.reader(csvfile, delimiter=",")
    header = next(csvreader)
    for row in csvreader:
        date = row[0]
        profit_loss = int(row[1])
        
        total_months += 1
        
        net_total_profit_losses += profit_loss
        
        if total_months > 1:
            change = profit_loss - previous_month_profit_loss
            monthly_changes.append(change)

            if change > greatest_increase[1]:
                greatest_increase[0] = date
                greatest_increase[1] = change
            if change < greatest_decrease[1]:
                greatest_decrease[0] = date
                greatest_decrease[1] = change
        previous_month_profit_loss = profit_loss
```

```python
average_change = sum(monthly_changes) / len(monthly_changes)
```

```python
print("Financial Analysis")
print("----------------------------")
print(f"Total Months: {total_months}")
print(f"Net Total Profit/Losses: ${round(net_total_profit_losses/1000000)}M")
print(f"Average Change: ${round(average_change/1000,2)}K")
print(f"Greatest Increase in Profits: {greatest_increase[0]} (${round(greatest_increase[1]/1000000,)}M)")
print(f"Greatest Decrease in Losses: {greatest_decrease[0]} (${round(greatest_decrease[1]/1000000,)}M)")
```

```python
with open("PYBank_Financial_Analysis.txt","w") as output_file:
    output_file.write("PYBank Financial Analysis/n")
    
    output_file.write("----------------------------")
   
    output_file.write(f"Total Months: {total_months}")
    
    output_file.write(f"Net Total Profit/Losses: ${round(net_total_profit_losses/1000000)}M")
    
    output_file.write(f"Average Change: ${round(average_change/1000,2)}K")
    
    output_file.write(f"Greatest Increase in Profits: {greatest_increase[0]} (${round(greatest_increase[1]/1000000,)}M)")
    
    output_file.write(f"Greatest Decrease in Losses: {greatest_decrease[0]} (${round(greatest_decrease[1]/1000000,)}M)")

print("Results written to PYBank_Financial_Analysis.txt")
```
